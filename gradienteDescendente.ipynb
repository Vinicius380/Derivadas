### Gradiente Descendeste

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
```

```python
# Função Custo
def cost_function(x,y):
    return x**2 + y**2

#Vetor Gradiente
def gradiente(x,y):
    return np.array([2*x,2*y])

#Gráfico
x_grid=np.linspace(-15,15,100)
y_grid=np.linspace(-15,15,100)
X,Y = np.meshgrid(x_grid,y_grid)
Z = cost_function(X,Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection = "3d")

ax.plot_surface(X,Y,Z, cmap="cool", alpha=0.8)

ax.set_title("Plot 3D funcao Z (x,y)", fontsize=14)
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.set_zlabel("z", fontsize=12)
```

```python
from matplotlib.animation import PillowWriter

#Parametros do Gradiente
learning_rate = 0.5
iterations = 50

#Starting Point 
x, y= random.randint(-10, 10), random.randint(-10,10)
x_vals, y_vals = [x],[y]

#Crete a grid for contour plot
x_grid=np.linspace(-15,15,100)
y_grid=np.linspace(-15,15,100)
X,Y = np.meshgrid(x_grid,y_grid)
Z = cost_function(X,Y)

#Countour chart
fig, ax = plt.subplots(figsize=(8,6))
contour = ax.contour(X,Y,Z, levels=20, cmap="viridis")
path, = ax.plot([], [], "ro-", label="Gradient Descent Path")
min_point = ax.scatter(0,0, color="blue", label="Minimum")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

#Function For Animation
def init():
    path.set_data([], [])
    return path,

def update(frame):
    global x,y
    grad=gradiente(x,y)
    x -= learning_rate * grad[0]
    y -= learning_rate * grad[1]
    x_vals.append(x)
    y_vals.append(y)
    path.set_data(x_vals, y_vals)
    return path,

ani = FuncAnimation(fig, update, frames=iterations, init_func=init,blit=True,repeat=False)
ani.save("gradiente_descent.gif", writer=PillowWriter(fps=10))
from IPython.display import Image
Image("gradiente_descent.gif")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation

# Função Custo
def cost_function(x,y):
    return x**2 + y**2 -2*x - 4*y

#Vetor Gradiente
def gradiente(x,y):
    return np.array([2*x - 2, 2*y - 4])

#Gráfico
x_grid=np.linspace(-15,15,100)
y_grid=np.linspace(-15,15,100)
X,Y = np.meshgrid(x_grid,y_grid)
Z = cost_function(X,Y)

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection = "3d")

ax.plot_surface(X,Y,Z, cmap="cool", alpha=0.8)

ax.set_title("Plot 3D funcao Z (x,y)", fontsize=14)
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.set_zlabel("z", fontsize=12)
```
